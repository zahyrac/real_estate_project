# Price comparison per neighborhood

```python
# Average price of sold properties before the pandemic by neighborhood
pernei = beforeCovid.groupby(["neighborhood","borough"]).price.mean().sort_values(ascending=False)
beforeCovid.groupby(["neighborhood","borough"]).price.mean().sort_values(ascending=False)
```

```python
b_pernei = pd.DataFrame(pernei).reset_index()
b_top5= b_pernei.head(5)
b_bottom5= b_pernei.tail(5)
b_top5
```

```python
fig, axes = plt.subplots(ncols=2, sharey=True)

sns.barplot(x='neighborhood', y='price', data=b_top5,  ax=axes[0]).set(title='top 5 before pandemic')
sns.barplot(x='neighborhood', y='price', data=b_bottom5, ax=axes[1]).set(title='bottom 5 before pandemic')
fig.set_size_inches(20, 7)
sns.set(font_scale = .7)
plt.show()

```

```python
# Average price of sold properties before the pandemic by neighborhood
pernei2 = duringCovid.groupby(["neighborhood","borough"]).price.mean().sort_values(ascending=False)
duringCovid.groupby(["neighborhood","borough"]).price.mean().sort_values(ascending=False)
```

```python
d_pernei = pd.DataFrame(pernei2).reset_index()
d_top5= d_pernei.head(5)
d_bottom5= d_pernei.tail(5)
```

```python
sns.set(font_scale = .7)
plt.show()

fig, axes = plt.subplots(ncols=2, sharey=True)

sns.barplot(x='neighborhood', y='price', data=d_top5,  ax=axes[0]).set(title='top 5 during pandemic')
sns.barplot(x='neighborhood', y='price', data=d_bottom5, ax=axes[1]).set(title='bottom 5 during pandemic')
fig.set_size_inches(20, 7)


```

```python

fig, axes = plt.subplots(2,2, sharey=True)
sns.barplot(x='neighborhood', y='price', data=b_top5,  ax=axes[0,0]).set(title='top 5 before pandemic')
sns.barplot(x='neighborhood', y='price', data=b_bottom5, ax=axes[0,1]).set(title='bottom 5 before pandemic')
sns.barplot(x='neighborhood', y='price', data=d_top5,  ax=axes[1,0]).set(title='top 5 during pandemic')
sns.barplot(x='neighborhood', y='price', data=d_bottom5, ax=axes[1,1]).set(title='bottom 5 during pandemic')

sns.set(font_scale = 0.8)
fig.set_size_inches(20, 15)
fig.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2,2, sharey=True)
sns.barplot(x='neighborhood', y='price',hue="borough", dodge=False, data=b_top5,  ax=axes[0,0]).set(title='top 5 before pandemic')
sns.barplot(x='neighborhood', y='price',hue="borough", dodge=False, data=b_bottom5, ax=axes[0,1]).set(title='bottom 5 before pandemic')
sns.barplot(x='neighborhood', y='price',hue="borough",dodge=False, data=d_top5,  ax=axes[1,0]).set(title='top 5 during pandemic')
sns.barplot(x='neighborhood', y='price',hue="borough",dodge=False, data=d_bottom5, ax=axes[1,1]).set(title='bottom 5 during pandemic')

sns.set(font_scale = 0.8)
fig.set_size_inches(20, 15)
fig.tight_layout()
plt.show()
```
